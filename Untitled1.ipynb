```python

def which_is_longer(feature_coords):

    num_layers = 0
    f0 = feature_coords
    while type(f0) == list:
        f0 = f0[0]
        num_layers +=1
        
    temp = feature_coords
    
    if num_layers > 3:
        for i in range(num_layers-3):
            processed_feature_coords = temp
            temp = temp[0]

    elif num_layers == 3:
        processed_feature_coords = temp
    
    else:
        processed_feature_coords = temp

    
    temp = 0
    longest = 0
    
    for long_idx, i_dont_know_why in enumerate(processed_feature_coords):
        while len(i_dont_know_why) == 1:
            i_dont_know_why = i_dont_know_why[0]

        if len(i_dont_know_why) > temp:
            temp = len(i_dont_know_why)
            longest = long_idx
    
    return longest


def process_features_coords(features):
    
    processed_coord_list = []
    
    assert type(features) == list
        
    for feature in features:
        #print(type(feature['geometry']['coordinates']))
        
        if type(feature['geometry']['coordinates']) == list:
            
            f_id = which_is_longer(feature['geometry']['coordinates'])

            processed_coord = feature['geometry']['coordinates'][f_id]
            
            num_layers = 0
            f0 = processed_coord
            while type(f0) == list:
                f0 = f0[0]
                num_layers +=1
                
            if num_layers == 3:

                processed_coord = processed_coord[0]
                
        processed_coord_list.append(processed_coord)
        
    return processed_coord_list
```

## gernerate and move Masks

```python
folder_list = os.listdir("./seperate_labeled/")
folder_list
```

```python
for folder_name in folder_list:
    #folder_name = "human1"

    root_folder = "./seperate_labeled/{}/".format(folder_name)

    channel_list = os.listdir(root_folder)



    img_size = (756,672)

    threshold = 0.6

    res_cell_list = []



    for idx, channel in enumerate(channel_list): ## each channel
        one_channel_result = np.zeros(img_size)
        cell_count = 1

        if channel != "useless":

            print("================="+channel+"=====================")
            ann_json_path = os.path.join(root_folder,channel,"annotations.json")
            with open(ann_json_path) as f:
                data_dict = json.load(f)

            list_of_point_list = process_features_coords(data_dict)
            print(len(list_of_point_list))

            tmp_ch_cell_list = [] ## record one ch cells that have small iou with res_cell_list. 

            for cell in list_of_point_list:

                add_new = True

                cell_points = np.array(cell).astype(int) ## each cell

                cell_mask = cv2.drawContours(np.zeros((img_size[0], img_size[1]), np.uint8), [cell_points], -1, 255, cv2.FILLED)

                for i in range(img_size[0]):
                    for j in range(img_size[1]):
                        if cell_mask[i][j] != 0:
                            one_channel_result[i][j] = cell_count

                cell_count += 1

                # cell_mask_contours, _ = cv2.findContours(mask_and, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

                cell_mask_area = cv2.contourArea(cell_points)

                if len(res_cell_list) == 0:
                    tmp_ch_cell_list.append(cell)

                else:

                    for checked_cell in res_cell_list:

                        res_points = np.array(checked_cell).astype(int) ## each cell

                        res_mask = cv2.drawContours(np.zeros((img_size[0], img_size[1]), np.uint8), [res_points], -1, 255, cv2.FILLED)

                        res_mask_area = cv2.contourArea(res_points)


                        mask_and = cv2.bitwise_and(res_mask, cell_mask)
                        mask_and_contours, _ = cv2.findContours(mask_and, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

                        if len(mask_and_contours)>0:
                            mask_and_area = cv2.contourArea(mask_and_contours[0])

                            new_cell_area_ratio = mask_and_area/cell_mask_area
                            checked_cell_area_ratio = mask_and_area/res_mask_area

                            if max(new_cell_area_ratio,checked_cell_area_ratio) > threshold:
                                if res_mask_area > cell_mask_area:
                                    add_new = False
                                else:
                                    res_cell_list.remove(checked_cell)

                    if add_new == True:
                        tmp_ch_cell_list.append(cell)
            #print("tmp_ch_cell_list = ",tmp_ch_cell_list) 
            res_cell_list.extend(tmp_ch_cell_list)
        np.save(root_folder+channel+"/"+channel+".npy", one_channel_result)
        one_channel_mask_img = Image.fromarray(one_channel_result).convert("I")
        one_channel_mask_img.save("./NTUH_train/{}_{}_masks.png".format(folder_name, channel))
# img_size = (756,672)

# test_mask = np.zeros((img_size[0], img_size[1]), np.uint8)

# for test_cell in res_cell_list:


# 	test_cell_points = np.array(test_cell).astype(int) ## each cell

# 	test_mask = cv2.drawContours(test_mask, [test_cell_points], -1, 255, cv2.FILLED)

# cv2.imwrite("{}.png".format(json_name),test_mask)
```

## after infer, put channels in same image back together

```python
import shutil
```

```python
infer_folder = "./NTUH_test/"
infer_imgs = [f for f in os.listdir(infer_folder) if "_masks" in f]
```

```python
infer_imgs
```

```python
for img_name in test_folder_list:
    organized_path = "./NTUH_test_organized/{}/".format(img_name)
    if not os.path.isdir(organized_path):
        os.mkdir(organized_path)

    for infer_img in infer_imgs:
        if img_name in infer_img:
            source_path = infer_folder + infer_img
            dest_path = organized_path + infer_img
            shutil.copyfile(source_path, dest_path)
```

## for each image, stack masks from each channel

```python
for img_name in test_folder_list:
    organized_path = "./NTUH_test_organized/{}/".format(img_name)
    if os.path.isdir(organized_path):
        channels = [f for f in os.listdir(organized_path) if not f.startswith('.')]
        blank_img = np.zeros_like(np.array(Image.open(organized_path+channels[0])))
        cell_count = 1
        for channel in channels:
            if "DAPI" not in channel and "Ch1Cy1" not in channel:
                img = np.array(Image.open(organized_path+channel))
                ## print("channel {} max {}".format(channel, np.max(img)))
                for i in range(1, np.max(img)+1):
                    if not np.any(np.logical_and(img == i, blank_img > 0)):
                        blank_img[img == i] = cell_count
                        cell_count += 1
        result_img = Image.fromarray(blank_img*255).convert("L")
        result_img.save("./NTUH_test_processed/{}.png".format(img_name))
```
