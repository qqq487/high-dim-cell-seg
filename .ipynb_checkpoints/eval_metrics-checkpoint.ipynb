```python
import cv2
import numpy as np
from cellpose import metrics
from scipy.ndimage import label
import matplotlib.pyplot as plt
import json
from PIL import Image
```

```python
pred = Image.open("../cellpose_sup/infer_results/44_multi_bce_dice_ks3/test_human4_masks.png")
```

```python
pred_array = (np.array(pred).astype("int32"))
```

```python
GT = np.load("../unet/train_data/multi_refined_mouse_human/human4_mask.npy").astype("int32")
```

```python
GT_labeled,_ = label(GT)
```

```python
pred_array.dtype
```

```python
GT_labeled.max()
```

```python
aji = metrics.aggregated_jaccard_index([GT_labeled],[pred_array])
```

```python
aji
```

```python
ap, tp, fp, fn = metrics.average_precision([GT_labeled], [pred_array], threshold=np.arange(0.5, 0.55, 0.05))

precision = tp / (tp + fp)
recall = tp / (tp + fn)

f1_score = tp / (2*tp + fn + fp)
```

```python
precision.mean()
```

```python
recall.mean()
```

```python
f1_score.mean()
```

```python
def plot_fig(pred, gt):
    # plt.imshow(gt)
    fig = plt.figure(figsize=(15,15))

    #f, axarr = plt.subplots(1,2)
    # Adds a subplot at the 1st position
    fig.add_subplot(1, 2, 1)

    # showing image
    plt.imshow(pred)
    plt.axis('off')
    plt.title("pred")

    # Adds a subplot at the 2nd position
    fig.add_subplot(1, 2, 2)

    # showing image
    plt.imshow(gt)
    plt.axis('off')
    plt.title("ground_truth")

    plt.show()
```

```python
pred = [np.load("./test_data/test_imgs_multi_with_nuclei/outputs/human4.npy").astype(np.int32)]
#gt = [(np.load("./test_data/test_imgs_multi_with_nuclei/outputs/23.npy")).astype(np.uint8)]

gt = [label(np.load("./train_data/multi_refined_mouse_human/human4_mask.npy"))[0]]

plot_fig(pred[0],gt[0])

#scales = np.arange(0.025, 0.275, 0.025)
#scales = np.arange(0.5, 0.955, 0.05)
#bdry_precion, bdry_recall, bdry_fscore = metrics.boundary_scores(pred,gt,scales)

aji = metrics.aggregated_jaccard_index(gt, pred)
ap, tp, fp, fn = metrics.average_precision(gt, pred, threshold=np.arange(0.5, 0.55, 0.05))

precision = tp / (tp + fp)
recall = tp / (tp + fn)

f1_score = tp / (2*tp + fn + fp)

```
